```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('work.csv')
df.head()
```

```python
df=df.iloc[:,2:] # to remove user id and gender column because they will not be req
df.head(5)
```

```python
# without standardization
for col in df.columns:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(df[col], shade=True, color='purple') # Kernel Density Estimation shows curve form of graph
    plt.title(f'KDE of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

```python
# mean of df without standardization
df.mean()
```

```python
# SD of df without standardization
df.std()
```

```python
# applying standardizatioin
standardized_df = (df - df.mean()) / df.std()
```

```python
standardized_df.mean() # means are approximately zero
```

```python
standardized_df.std() # standard deviation has become 1
```

```python
# after standardization
for col in standardized_df.columns:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(df[col], shade=True, color='skyblue') # Kernel Density Estimation shows curve form of graph
    plt.title(f'KDE of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

```python
# creating datasrt which will show difference for better understanding
np.random.seed(42)
normal_data = np.random.normal(50, 10, 1000)
skewed_data = np.random.exponential(20, 1000)
bimodal_data = np.concatenate([
    np.random.normal(30, 5, 500),
    np.random.normal(70, 5, 500)
])
df1=pd.DataFrame({
    'Normal': normal_data,
    'Skewed': skewed_data,
    'Bimodal': bimodal_data
})
```

```python
df1.head()
```

```python
df1.mean()
```

```python
df1.std()
```

```python
# before standardization
for col in df1.columns:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(df[col], shade=True, color='skyblue') # Kernel Density Estimation shows curve form of graph
    plt.title(f'KDE of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

```python
# applying standardization using built in library
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
standardized_data=scaler.fit_transform(df)
standardized_df=pd.DataFrame(standardized_data, columns=df1.columns)
```

```python
# after standardization
for col in standardized_df.columns:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(df[col], shade=True, color='red') # Kernel Density Estimation shows curve form of graph
    plt.title(f'KDE of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

```python
standardized_df.mean()
```

```python
standardized_df.std()
```
